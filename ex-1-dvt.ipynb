```python
import pandas as pd

df = pd.read_csv('/kaggle/input/datasets/rohitsahoo/sales-forecasting/train.csv')

# 2. Basic checking
print("Shape:", df.shape)
print(df.head())
print(df.info())

# 3. Check missing values
print(df.isnull().sum())

# 4. Handle missing Postal Code
df["Postal Code"] = df["Postal Code"].fillna(0)

# 5. Convert Postal Code to integer
df["Postal Code"] = df["Postal Code"].astype(int)

# 6. Remove duplicates
df = df.drop_duplicates()

# 7. Convert date columns
df["Order Date"] = pd.to_datetime(df["Order Date"], errors="coerce")
df["Ship Date"] = pd.to_datetime(df["Ship Date"], errors="coerce")

# 8. Create new useful date columns
df["Order Year"] = df["Order Date"].dt.year
df["Order Month"] = df["Order Date"].dt.month
df["Order Month Name"] = df["Order Date"].dt.month_name()
df["Order Quarter"] = df["Order Date"].dt.quarter

# 9. Create shipping duration column
df["Shipping Days"] = (df["Ship Date"] - df["Order Date"]).dt.days

# 10. Clean text columns
text_columns = [
    "Ship Mode", "Customer Name", "Segment", "Country",
    "City", "State", "Region", "Category",
    "Sub-Category", "Product Name"
]

for col in text_columns:
    df[col] = df[col].astype(str).str.strip()

# 11. Check sales column
df["Sales"] = pd.to_numeric(df["Sales"], errors="coerce")

# 12. Remove rows with invalid dates or sales
df = df.dropna(subset=["Order Date", "Ship Date", "Sales"])

# 13. Final check
print("Final Shape:", df.shape)
print(df.info())
print(df.describe())

# 14. Save cleaned dataset
df.to_csv("superstore_cleaned.csv", index=False)

print("Preprocessing completed successfully.")
```
